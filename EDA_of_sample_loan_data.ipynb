2.	loan_user_id: - unique loan user identifier

4.	loan_repay_date: - user loan repayment date

6.	loan_status: - status of the loan
a.	loans_status=2 -> loan got paid by the user
b.	loan_status=7-> loan got defaulted
7.	laon_created_at: - loan given time

```python
# define numerical & categorical columns
numeric_features = [feature for feature in df2.columns if df2[feature].dtype != 'O']
categorical_features = [feature for feature in df2.columns if df2[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```
