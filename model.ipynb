## Data Preprocessing

profile_birth_date has total 2261 missing value and for profile_city we have total 2435 missing values.hence we drop this columns

```python
columns_to_drop = ['profile_birth_date', 'profile_city','loan_repay_extended_date']
df.drop(columns_to_drop, axis=1, inplace=True)
```

for profile_residence_state column has 25 missing values.

```python
# Calculate the mode of the 'Category' column
mode_category = df['profile_residence_state'].mode()[0]

# Fill missing values in the 'Category' column with the mode
df['profile_residence_state'].fillna(mode_category, inplace=True)
```

```python
# Dropping rows with missing values for specific columns permanently
columns_to_check = ['trans_name', 'trans_personal_category','trans_personal_cat_details']  # Specify columns to check for missing values
df.dropna(subset=columns_to_check, inplace=True)  # Drop rows with missing values for specified columns
```

```python
df.isna().sum()
```

#Removing outliers

trans_amount,trans_after_balance columns has outliers.

```python
# Calculate the mean of the 'trans_amount' column
mean_numeric_column = df['trans_amount'].mean()
mean_numeric_column
# Fill missing values in the 'Numeric_Column' column with the mean
df['trans_amount'].fillna(mean_numeric_column, inplace=True)
```

```python
mean_numeric_column1 = df['trans_after_balance'].mean()
mean_numeric_column1
# Fill missing values in the 'Numeric_Column' column with the mean
df['trans_after_balance'].fillna(mean_numeric_column1, inplace=True)
```

```python
# Find out multiple categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Displaying the list of categorical columns
print("Categorical columns:", list(categorical_columns))
```

```python
# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply LabelEncoder to multiple categorical columns and overwrite the existing columns
for column in ['profile_residence_state', 'trans_name', 'trans_personal_category', 'trans_personal_cat_details']:
    df[column] = label_encoder.fit_transform(df[column])

# Display the encoded DataFrame
df.head()
```

```python
# Initialize OneHotEncoder
onehot_encoder = OneHotEncoder(sparse=False)

# Apply OneHotEncoder to the 'Category' column and overwrite the original column
encoded_categories = onehot_encoder.fit_transform(df[['profile_registerd_device']])
encoded_df = pd.DataFrame(encoded_categories, columns=onehot_encoder.get_feature_names_out(['profile_registerd_device']))
df.drop('profile_registerd_device', axis=1, inplace=True)  # Drop the original column
df[encoded_df.columns] = encoded_df  # Assign the encoded values to the original column names
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
target=df1['loan_status']
data = df1.drop(columns=['loan_status'])
# Extract numeric columns for scaling
numeric_columns = data.select_dtypes(include=['number']).columns

# Initialize StandardScaler
scaler = StandardScaler()

# Scale only the numeric columns
data[numeric_columns] = scaler.fit_transform(data[numeric_columns])

df1 = pd.concat([data, target], axis=1)
# Display (the scaled DataFrame
df1.head()
```

```python
#df.to_csv('scaled_data.csv', index=False)
```

```python
X = df1.drop(columns=['loan_status'])
y = df1['loan_status']
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.25,
                                                    random_state=7)
print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

# Feature selection

```python
from sklearn.feature_selection import SelectKBest, f_classif
# Perform feature selection
# SelectKBest with f_classif score function (suitable for classification problems)
selector = SelectKBest(score_func=f_classif, k=10)  # Select top 12 features
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# Now, X_train_selected and X_test_selected contain only the selected features

# Initialize and train a classification model using the selected features
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_selected, y_train)

# Make predictions on the test set using the model with selected features
y_pred = clf.predict(X_test_selected)
```

```python
# Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train_selected, y_train)
log_reg_pred = log_reg.predict(X_test_selected)
log_reg_accuracy = accuracy_score(y_test, log_reg_pred)
print("Logistic Regression Accuracy:", log_reg_accuracy)

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train_selected, y_train)
rf_pred = rf.predict(X_test_selected)
rf_accuracy = accuracy_score(y_test, rf_pred)
print("Random Forest Accuracy:", rf_accuracy)


# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train_selected, y_train)
dt_pred = dt.predict(X_test_selected)
dt_accuracy = accuracy_score(y_test, dt_pred)
print("Decision Tree Accuracy:", dt_accuracy)

# XGBoost
xgb_classifier = xgb.XGBClassifier()
xgb_classifier.fit(X_train_selected, y_train)
xgb_pred = xgb_classifier.predict(X_test_selected)
xgb_accuracy = accuracy_score(y_test, xgb_pred)
print("XGBoost Accuracy:", xgb_accuracy)

# Support Vector Machine
svm = SVC()
svm.fit(X_train_selected, y_train)
svm_pred = svm.predict(X_test_selected)
svm_accuracy = accuracy_score(y_test, svm_pred)
print("SVM Accuracy:", svm_accuracy)
```
