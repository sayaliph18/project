```python
print("Categories in 'profile_city' variable:     ",end=" " )
print(df1['profile_city'].unique())
print('Total count: ',df1['profile_city'].nunique())
print('------------------------------------------------------------------')

print("Categories in 'profile_residence_state' variable:     ",end=" " )
print(df1['profile_residence_state'].unique())
print('Total count: ',df1['profile_residence_state'].nunique())
print('------------------------------------------------------------------')

print("Categories in 'profile_registerd_device' variable:     ",end=" " )
print(df1['profile_registerd_device'].unique())
print('Total count: ',df1['profile_registerd_device'].nunique())
print('------------------------------------------------------------------')

```

```python
# define numerical & categorical columns
numeric_features = [feature for feature in df1.columns if df1[feature].dtype != 'O']
categorical_features = [feature for feature in df1.columns if df1[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```
